<h1>Differentialrechnung</h1>

<h2>B1</h2>

$y:=(5x^3-4x)(x^2+5x)$

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.misc import derivative
```

```python
h=1e-6 # Schrittweite
x1,x2=-2,2 
y1,y2=-5,2
plt.xlim(x1,x2)
plt.ylim(y1,y2)

def f(x):
	return (5*x**3-4*x)*(x**2+5*x)

def df_1(x):
	return derivative(f,x,dx=h)

def df_2(x):
	return derivative(df_1,x,dx=h)

def df_3(x):
	return derivative(df_2,x,dx=h)

x = np.linspace(x1, x2, 500)
plt.plot(x, f(x), lw=2, color='r')
plt.plot(x, df_1(x), 'b--', label='1. Ableitung')
plt.plot(x, df_2(x), 'g--', label='2. Ableitung')
#plt.plot(x, df_3(x), 'b--', label='3. Ableitung')
plt.legend(loc='best')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```

<h2>B2</h2>

$y:=x^5 \cdot \ln x$

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.misc import derivative

h=1e-6 # Schrittweite
x1,x2=-0.5,1.5 
y1,y2=-1,3
plt.xlim(x1,x2)
plt.ylim(y1,y2)

def f(x):
	return x**5*np.log(x)

def df_1(x):
	return derivative(f,x,dx=h)

def df_2(x):
	return derivative(df_1,x,dx=h)

def df_3(x):
	return derivative(df_2,x,dx=h)

x = np.linspace(x1, x2, 500)
plt.plot(x, f(x), lw=2, color='r')
plt.plot(x, df_1(x), 'b--', label='1. Ableitung')
plt.plot(x, df_2(x), 'g--', label='2. Ableitung')
plt.plot(x, df_3(x), 'y--', label='3. Ableitung')
plt.legend(loc='best')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```
